```python
N = 5015
F = 100000.0
f_step = F/N
f_bin = np.arange(0.0, F, f_step)

#def correlator(f_mag, f_rms, f_rans):
for i in ilist:
    os.chdir(path_mag)
    mag = pd.read_csv(maglist[i]).set_index('nodenumber')
    awf = mag.apply(lambda x: np.sum(x*f_bin[:1254])/np.sum(x), axis=1) #amplitude weighted frequency
    del mag
    awf
    corr = pd.DataFrame({'awf': awf})
    del awf
    os.chdir(path_rms)
    rms = pd.read_csv(rmslist[i]).set_index('nodenumber')
    corr['sound-pressure'] = rms['sound-pressure']
    #corr['sound-intensity'] = rms['sound-intensity']
    del rms
    os.chdir(path_rans)
    rans = pd.read_csv(ranslist[i]).set_index('nodenumber')
    rans.columns = rans.columns.str.strip()
    corr['vorticity-mag'] = rans['vorticity-mag']
    corr['turb-kinetic-energy'] = rans['turb-kinetic-energy']
    corr['viscosity-turb'] = rans['viscosity-turb']
    corr['specific-diss-rate'] = rans['specific-diss-rate']
    corr['turb-intensity'] = rans['turb-intensity']
    corr = corr.rename_axis(str(rmslist[i])[:10], axis=1)
    del rans
    result = corr.corr(method='pearson', min_periods=1).iloc[2:, :2]
    del corr
    print(result)
    #return result
```
