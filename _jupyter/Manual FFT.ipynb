```python
def manual_fft(x):
    X = []
    N = 5015
    n_list = np.arange(1,5016) #number of samples
    k_list = np.arange(0,5015) #number of frequency bins
    X = np.fft.fft(x)
    for k in k_list:
        b_kn = (2*np.pi*k*n_list)/N
        X_k = np.sum(x*(np.cos(-b_kn) + 1j*np.sin(-b_kn)))
        X = X.append(X_k)
    return np.concatenate([X])
```
