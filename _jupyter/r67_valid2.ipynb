```python
location = 'int-12'
os.chdir(path_valid)
coords = pd.read_csv('rans-int-12.dat', delimiter=",", header=0, skiprows=0, decimal='.').set_index('nodenumber')
x = coords['    x-coordinate']
y = coords['    y-coordinate']
z = coords['    z-coordinate']
mach = coords[' rel-mach-number']

mach_plot, ax0 = plt.subplots(1, 1, figsize=(6, 6), dpi=150)
m_plot = ax0.scatter(z, x, c=mach, s=0.5, cmap=plt.cm.jet, vmin=0, vmax=1.65)
mach_plot.colorbar(m_plot, ax=ax0, orientation="vertical", pad=0.1)
ax0.set_title(str('RANS relative Mach number at 10% span'))
ax0.set_aspect('equal', 'box')
ax0.set_xlim(-0.05, 0.150)
ax0.set_ylim(-0.08, 0.150)
ax0.set_xticks(np.arange(-0.05, 0.150, 0.025))
ax0.set_yticks(np.arange(-0.075, 0.150, 0.025))
ax0.set_xlabel('z - coordinate')
ax0.set_ylabel('x - coordinate')
plt.grid()
plt.show()
#os.chdir(path_plots)
#plt.savefig(str(str(location) + '-rms-spl.png'), bbox_inches='tight')
#plt.close()
```
